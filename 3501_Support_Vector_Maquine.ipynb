##### Representación de limite de decision 

```python
def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    D = svm_clf.intercept_[0]

    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = - w [0] / w [1] * x0 - D / w [1]

    margin = 1 / w [1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    # Esto es la nube de color rojo que se tiene que visualizar la frente de todas las representaciones.
    
    svs = svm_clf.support_vectors_ # Definimos los datos de entrenamiento más cercanos a la frontera de decisión.
    plt.scatter(svs[:, 0], svs[:, 1], s = 180, facecolors = '#FFAAAA') # Gráficación de los vectores de soporte 
    plt.plot(x0, decision_boundary, "k-", linewidth = 2) # Dibujar la frontera de decisión
    plt.plot(x0, gutter_up, "k--", linewidth = 2) # Margen superior
    plt.plot(x0, gutter_down, "k--", linewidth = 2) # Margen inferior
```

```python
plt.figure(figsize = (12, 6))
plt.plot(X_train_reduced.values[:, 0] [y_train == "phishing"], X_train_reduced.values[:, 1][y_train == "phishing"], "g^")
plt.plot(X_train_reduced.values[:, 0] [y_train == "benign"], X_train_reduced.values[:, 1][y_train == "benign"], "bs")
plot_svc_decision_boundary(svm_clf, 0, 1) # Frontera de decisión. (Guiones de color negro)
# Personalización de la impresión
plt.title("$C = {}$".format(svm_clf.C), fontsize = 16)
plt.axis([0, 1, -100, 250])
plt.xlabel("domainUrlRatio", fontsize = 13)
plt.ylabel("domainUrllength", fontsize = 13)
plt.show()
```

#### TAREA:
##### Mostrar los valores de color rojo en la parte de enfrente.

El parámetro `zorder` controla el orden en que se dibujan los parametos del gráfico.
- Elementos con valores **más bajos** de zorder se dibujan primero (quedan al fondo).
- Elementos con valores **más altos** se dibujan después (quedan al frente).

```python
def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    D = svm_clf.intercept_[0]

    # Calcular los puntos de la frontera y márgenes
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = - w[0] / w[1] * x0 - D / w[1]
    margin = 1 / w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    # Obtener vectores de soporte
    svs = svm_clf.support_vectors_
    
    # Graficar vectores de soporte (fondo)
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA', edgecolors='k', zorder=3, label="Support Vectors")
    
    # Graficar la frontera de decisión y márgenes (frente de los vectores de soporte)
    plt.plot(x0, decision_boundary, "k-", linewidth=2, zorder=4, label="Decision Boundary")
    plt.plot(x0, gutter_up, "k--", linewidth=2, zorder=4, label="Margins")
    plt.plot(x0, gutter_down, "k--", linewidth=2, zorder=4)

    # Agregar leyenda
    plt.legend()
```

```python
plt.figure(figsize = (12, 6))
plt.plot(X_train_reduced.values[:, 0] [y_train == "phishing"], X_train_reduced.values[:, 1][y_train == "phishing"], "g^", label="Phishing", zorder=3)
plt.plot(X_train_reduced.values[:, 0] [y_train == "benign"], X_train_reduced.values[:, 1][y_train == "benign"], "bs", label="Benign", zorder=3)
plot_svc_decision_boundary(svm_clf, 0, 1) # Frontera de decisión. (Guiones de color negro)
# Personalización de la impresión
plt.title("$C = {}$".format(svm_clf.C), fontsize = 16)
plt.axis([0, 1, -100, 250])
plt.xlabel("domainUrlRatio", fontsize = 13)
plt.ylabel("domainUrllength", fontsize = 13)
plt.show()
```

## 6.- SVM: Kernel No Lineal

#### Entrenamiento del algoritmo con un DataSet reducido

```python
# Para representar el limite de decisión, se tiene que pasar la variable a numerica 
y_train_num = y_train.factorize()[0]
y_val_num = y_val.factorize()[0]
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import LinearSVC

polynomial_svm_clf = Pipeline([
    ("poly_features", PolynomialFeatures(degree=3)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC(C=20, loss="hinge", random_state=42, max_iter=100000))
    ])

polynomial_svm_clf.fit(X_train_reduced, y_train_num)
```

### Represntación del limite de decisión

```python
def plot_dataset(X, y):
    plt.plot(X[:,0][y==1],X[:,1][y==1], "g.")
    plt.plot(X[:,0][y==0], X[:,1][y==0], "b.")
```

```python
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()] #REDIMENSINAR
    y_pred = clf.predict(X).reshape(x0.shape) # Hacemso al prediccion en Y a partir de X
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap = plt.cm.brg, alpha=0.2) # Mapa de contorno
    plt.contourf(x0, x1, y_decision, cmap = plt.cm.brg, alpha=0.2)

fig, axes = plt.subplots(ncols=2, figsize=(15, 5), sharex=True)
plt.sca(axes[0])
plot_dataset(X_train_reduced.values, y_train_num)
plot_predictions(polynomial_svm_clf, [0,1, -100, 250])
plt.xlabel("domainUrlRatio", fontsize=11)
plt.ylabel("domainlength", fontsize=11)
plt.sca(axes[1])
plot_predictions(polynomial_svm_clf, [0, 1, -100, 250])
plt.xlabel("domainlength", fontsize=11)
plt.ylabel("domainlength", fontsize=11)
plt.show()
```
